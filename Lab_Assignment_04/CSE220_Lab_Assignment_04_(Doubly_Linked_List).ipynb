```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
  # Creates a Non Dummy Headed Circular Doubly Linked List using the values from the given array a.
    if type(a) == list:
        self.head = Node(a[0], None, None)
        tail = self.head
        for i in range(1, len(a)):
            n = Node(a[i], None, None)
            tail.next = n
            n.prev = tail
            tail = tail.next
            self.head.prev = n
            tail.next = self.head
            tail = n
    else:
        self.head = a
  
  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    count = 0
    temp = self.head
    while True:
        count += 1
        temp = temp.prev
        if temp == self.head:
            break
    return count 
  
  # prints the elements in the list
  def forwardprint(self):
    temp = self.head
    while temp != self.head.prev:
        print(temp.element, end = ', ')
        temp = temp.next
        if temp == self.head.prev:
            print(f"{self.head.prev.element}.")
            break
        
  # prints the elements in the list backward
  def backwardprint(self):
    temp = self.head
    while True:
        print(temp.prev.element, end = ', ')
        temp = temp.prev
        if temp.prev == self.head:
            print(f"{self.head.element}.")
            break

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    count = 0
    if idx > 0 and idx < self.countNode():
        tail = self.head
        while True:
            if count == idx:
                return tail
            tail = tail.next
            count +=1
    else:
        return None
  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    temp = self.head.next
    count = 1
    while temp != self.head:
        if elem == temp.element:
            return count
        count += 1
        temp = temp.next

    if elem == temp.element:
       count = 0
       return count 
    
    return -1


  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    n1 = Node(elem, None, None)

    if idx < 0 or idx > self.countNode():
       return None
    
    elif idx == 0:
        k = self.head.prev
        n1.next = self.head
        self.head.prev = n1
        n1.prev = k
        k.next = n1
        self.head = n1

    else:
        k = self.nodeAt(idx-1)
        temp  = k.next
        k.next = n1
        n1.next = temp
        temp.prev = n1
        n1.prev = k
       

  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    temp = self.head
    if idx == 0:
        removed_elem = self.head
        self.head = self.head.next
        removed_elem.prev.next = self.head
        self.head.prev = removed_elem.prev
        removed_elem.prev = None
        removed_elem.next = None
        return str(removed_elem.element)
    elif idx < self.countNode():
        n1 = self.nodeAt(idx-1)
        n2 = n1.next.next
        removed_elem = self.nodeAt(idx)
        removed_elem.prev = None
        removed_elem.next = None
        n1.next = n2
        n2.prev = n1
        return str(removed_elem.element)
    else:
       return None
```
