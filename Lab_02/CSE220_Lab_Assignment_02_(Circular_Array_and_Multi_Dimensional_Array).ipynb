```python
text1 = input()
text2 = input()

if len(text1)>10 or len(text2)>10:
    print("Invalid Input Size")

else:
    print("Your inpur is valid. Press any key and then enter to continue!!!")
    arr = [[0] * 10, [0] * 10]
    # Constructing the arr[0]
    for i in range(0, len(text1)):
        arr[0][i] = text1[i]
        if 65 <= ord(arr[0][i]) <= 90:
            start_idx1 = i
            x = arr[0][i]
            shiftleft = start_idx1

            print(f"Top Board Start Character: {x}")
            print(f"Top Board Start Index: {start_idx1}")


    # Shifting left with respect to uppercase letter
    temp = [0] * shiftleft
    for i in range(0,shiftleft):
      temp[i] =  arr[0][i]
    for i in range(0, len( arr[0]) - shiftleft):
       arr[0][i] =  arr[0][i+shiftleft]
    for i in range(len( arr[0])-1,(len( arr[0]) - shiftleft)-1,-1):
       arr[0][i] = temp[i-(len( arr[0]) - shiftleft)]


    # Constructing the arr[1]
    for j in range(0, len(text2)):
        arr[1][j] = text2[j]
        if 65 <= ord(arr[1][j]) <= 90:
            start_idx2 = j
            y = arr[1][j]
            shiftRightt = (len(arr[1])-1) - start_idx2

            print(f"Top Board Start Character: {y}")
            print(f"Top Board Start Index: {start_idx2}")

    # Shifting right with respect to uppercase letter
    temp = [0] * shiftRightt
    for i in range(len(arr[1])-1,(len(arr[1])-shiftRightt)-1,-1):
      temp[i-(len(arr[1])-shiftRightt)] = arr[1][i]

    for i in range(len(arr[1]) - 1,shiftRightt-1,-1):
      arr[1][i] = arr[1][i-shiftRightt]

    for i in range(0,shiftRightt):
        arr[1][i] = temp[i]

    print(arr)


    # Implementing a 2D Electronic Text Billboard
    while True:
        user_prompt = input("Enter any key and then enter to continue!!!")
        if user_prompt == "Q" or user_prompt == "q":
            print(f"Top board stat character: {x}")
            print(f"Top board stat index: {start_idx1}")
            print(f"Top board stat character: {y}")
            print(f"Top board stat index: {start_idx2}")
            break
        else:
            temp1 = arr[0][0]
            for i in range(0,len(text1)-1):
                arr[0][i] = arr[0][i+1]
            arr[0][len(arr[0])-1] = temp1
            temp2 = arr[1][len(arr[1])-1]
            for i in range(len(text2)-1,0,-1):
                arr[1][i] = arr[1][i-1]
            arr[1][0] = temp2

            for i in range(0,len(arr[0])):
                print(arr[0][i],end="")

            print()

            for j in range(0,len(arr[1])):
                print(arr[1][j],end="")

            print()

# inputs ⬇️
# giRtfel2th
# rightLeft2
```
